```python
#Import the Pandas and Numpy libraries.

import pandas as pd
import numpy as pd




# Loaded variable 'df' from URI: c:\Windows\System32\code\Data-Science-Python\titanic3.csv
df = pd.read_csv(r'c:\Windows\System32\code\Data-Science-Python\titanic3.csv')

def clean_data(df):
    # Replace missing values with the mean of each column in: 'age'
    df = df.fillna({'age': int(df['age'].mean())})
    df['age'] = df['age'] .astype('int64')
    # Replace missing values with the mean of each column in: 'fare'
    df = df.fillna({'fare': df['fare'].mean()})
    return df


#Create a new Dataframe the cleaned data
df_clean = clean_data(df.copy())
df_clean.head()

#Import the Seaborn and Matplotlib libraries
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=5, figsize=(30,5))
sns.set_style("whitegrid")
sns.violinplot(x="survived", y="age", hue="sex", data=df_clean, palette="bright",inner="quartile", ax=axs[0])
sns.pointplot(x="sibsp", y="survived", hue="sex", data=df_clean,palette="bright", ax=axs[1])
sns.pointplot(x="parch", y="survived", hue="sex", data=df_clean,palette="bright", ax=axs[2])
sns.pointplot(x="pclass", y="survived", hue="sex", data=df_clean,palette="bright", ax=axs[3])
sns.violinplot(x="survived", y="fare", hue="sex", data=df_clean,palette="bright",inner="quartile", ax=axs[4])

 
```
