# 2. Data loading and Preprocessing


```python
# Load the data
train_data = pd.read_csv('/content/train.csv')
test_data = pd.read_csv('/content/test.csv')

# Encoding the target variable
train_data['label'] = train_data['label'].map({'B': 0, 'M': 1})

# Handle NaN values if necessary (imputation or dropping)
train_data.dropna(subset=['label'], inplace=True)

# Prepare the data
X = train_data.drop(['id', 'label'], axis=1)
y = train_data['label']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Scaling the Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# 3.  Training Modeles(i.e, Perceptron, Logestic Regression, SVM, Decesion Trees, KNN and Random Forest) via scikit-learn


```python
from sklearn.linear_model import Perceptron
# Create and train the Perceptron model
perceptron = Perceptron(eta0=0.5, max_iter=1000, random_state=42)
perceptron.fit(X_train, y_train)

# Make predictions on the test set
y_pred = perceptron.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
# Create and train the Logistic Regression model
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logistic_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.svm import SVC
# Create and train the SVM model
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create and train the Decision Tree model
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = tree_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Create and train the KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```
